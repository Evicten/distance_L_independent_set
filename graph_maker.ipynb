```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import string
```

```python
# Create a graph
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (1, 4), (3, 5), (3, 6), (4, 7), (5, 8), (5, 9), (7, 10), (10, 4), (7,6)])


mis = set()
mis.add(5)
mis.add(7)
mis.add(2)
mis.add
# Draw the graph with nodes filled in the maximum independent set
pos = nx.spring_layout(G)  # positions for all nodes
node_labels = {node: string.ascii_lowercase[node-1] for node in G.nodes()}  
nx.draw(G, pos, with_labels=False, labels = node_labels, node_color=['lightcoral' if node in mis else 'beige' for node in G.nodes()], node_size=500, edgecolors = 'black', linewidths = 1)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 4))

# Create a new graph
G = nx.Graph()

# Add factor nodes (squares)
G.add_node('i', node_type='factor')
G.add_node('j', node_type='factor')
G.add_node('k', node_type='factor')
G.add_node('l', node_type='factor')

# Add variable nodes (circles)
G.add_node('v1', node_type='variable')
G.add_node('v2', node_type='variable')
G.add_node('v3', node_type='variable')

# Add edges between factor nodes and variable nodes
factor_nodes = ['i', 'j', 'k', 'l']
variable_nodes = ['v1', 'v2', 'v3']
G.add_edge('i', 'v1')
G.add_edge('i', 'v2')
G.add_edge('i', 'v3')
G.add_edge('j', 'v1')
G.add_edge('k', 'v2')
G.add_edge('l', 'v3')

pos = {'i': (0, 0), 'j': (0, -2), 'k': (1, 2), 'l': (-1, 2),
       'v1': (0, -1), 'v2': (0.5, 1), 'v3': (-0.5, 1)}

# Draw variable nodes
nx.draw_networkx_nodes(G, pos, nodelist=variable_nodes, node_color='lightblue', node_shape='o', node_size=500, edgecolors='black', linewidths=1)

# Draw factor nodes
nx.draw_networkx_nodes(G, pos, nodelist=factor_nodes, node_color='white', node_shape='s', node_size=500, edgecolors='black', linewidths=2)

# Draw labels for factor nodes
nx.draw_networkx_labels(G, pos, labels={'i': 'i', 'j': 'j', 'k': 'k', 'l': 'l'}, font_size=12)

# Draw edges
nx.draw_networkx_edges(G, pos)

# Remove axis
plt.axis('off')

# Show the plot
plt.show()

```
