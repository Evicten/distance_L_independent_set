```python
def initialize_psi(degree):
    shape = tuple([2] * degree + [2] * degree)  # Create shape dynamically
    psi = np.zeros(shape)
    return psi

def iterate_indices(indices, shape):
    if len(indices) == len(shape):
        # Base case: All indices are filled, yield the tuple
        yield tuple(indices)
    else:
        # Recursive case: Fill the next index
        for i in range(shape[len(indices)]):
            indices.append(i)
            yield from iterate_indices(indices, shape)
            indices.pop()  # Backtrack

# Example function using generalized indices
def example_function(shape):
    for indices in iterate_indices([], shape):
        print(indices[5])

degree = 4  # Example degree
shape = tuple([2] * degree + [2] * degree)
example_function(shape)
```
